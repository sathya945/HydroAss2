```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('./samples.csv')
```

```python
print(data.head())
print(data.columns)
```

```python
print(data.keys()[0])
new_data = data[data.keys()[0]]
columns = data.keys()[0].split(';')
```

```python
print(len(new_data[0].split(';')))
print(len(columns))
print(new_data[0].split(';'))
print(columns)
print(new_data)
```

```python
final_data = {}
for i in range(len(columns)):
    final_data[columns[i]] = []
for i in range(len(new_data)):
    now = new_data[i].split(';')
    for i in range(len(columns)):
        final_data[columns[i]].append(now[i]) 
```

```python
print(final_data.keys())
```

```python
df = pd.DataFrame.from_dict(final_data)
```

```python
print(df.columns)
print(df['Depth'])
```

```python
for each in df.columns:
    print(each)
    print(df[each].unique())
```

```python
print(df.isnull().sum())
```

```python
#removed all the rows with Data Quality Poor or Suspect, and also if they have any value flags

df.drop( df[ df['Data Quality'] == "Poor"].index , inplace=True)
df.drop( df[ df['Data Quality'] == "Suspect"].index , inplace=True)
df.drop( df[ df['Value Flags'] == "<" ].index , inplace=True)
df.drop( df[ df['Value Flags'] == ">" ].index , inplace=True)
```

```python
print(df.loc[(df['Sample Date'].str.match(pat = '1993')) & (df['GEMS Station Number'] == "IND00001")])

# take all the parameters so that we can create columns for our data

create_cols = df[df.columns[4]].unique()
```

```python
ans = {}
param_dict = {}
ans['GEMS Station Number'] = []
ans['Year'] = []
a = 2
for each in create_cols:
    ans[each] = []
    param_dict[each] = a
    a += 1

#iterating over all stations
for i in range(1,73):
    station = "IND000"
    if i < 10:
        station += "0"
    station += str(i)
    
    #iterating over all years
    for j in range(1979,2009):
        tmp = []
        cnt = []
        
        gem = df.loc[(df['Sample Date'].str.match(pat = str(j))) & (df['GEMS Station Number'] == station)]
        
        #creating a row initially
        tmp.append(station)
        tmp.append(j)
        cnt.append(0)
        cnt.append(0)
        for each in create_cols:
            tmp.append(None)
            cnt.append(0)
            
        #iterating over rows of each specific year and station
        for index, row in gem.iterrows():
            if tmp[param_dict[row['Parameter Code']]] is None:
                tmp[param_dict[row['Parameter Code']]] = float(row['Value'])
            else:
                tmp[param_dict[row['Parameter Code']]] += float(row['Value'])
            cnt[param_dict[row['Parameter Code']]] += 1
        
        #averaging over months for each year data
        for k in range(2,len(create_cols)+2):
            if tmp[k] is not None:
                tmp[k] = tmp[k]/cnt[k]
        
        #filling up entries for a row
        ans['GEMS Station Number'].append(tmp[0])
        ans['Year'].append(tmp[1])
        for each in create_cols:
            ans[each].append(tmp[param_dict[each]])
        print(i,j)
```

```python
final_df = pd.DataFrame.from_dict(ans)
```

```python
print(final_df.isnull().sum())
print(final_df)
```

```python
remove_cols = final_df.isnull().sum()
for each in remove_cols.keys():
    # removing columns if there are less than half values filled
    if remove_cols[each] > 1116:
        del final_df[each]
```

```python
print(final_df.isnull().sum())
print(final_df)
```

```python
#get average data yearly for all parameters at a particular station

average_df = final_df.groupby('GEMS Station Number').mean()
```

```python
average_df
```

```python
#iterate over each row in final data and replace the none values with average values

for index, row in final_df.iterrows():
    check_null = row.isnull()
#     print(row.isnull())
    for each in final_df.columns:
        if check_null[each] == True:
            ind = int(row['GEMS Station Number'][6:8])
            ind -= 1
            final_df.at[index, each] = average_df.iloc[ind][each]
```

```python
print(final_df.isnull().sum())
print(final_df)
```

```python
final_df.to_csv('preprocessed.csv')
```
